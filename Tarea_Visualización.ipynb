```python
import pandas as pd
import numpy as np
%matplotlib inline 
import matplotlib as mpl
import matplotlib.pyplot as plt
import datetime
from IPython.display import display, HTML
```

```python
df_chi = pd.read_excel('MigrantesChile (2005-2016).xlsx', sheet_name='Migrantes', skipfooter=3)
df_chi.head(3)
```

```python
df_chi['Total'] = df_chi.sum(axis=1) # Agrego columna total con la suma de emigrantes en el periodo 2005-2016
df_chi.drop(['ID Continent','ID Country'], axis=1, inplace=True)# Elimino ID Continent y ID Country
df_chi.head()
```

```python
df_chi.set_index('Country', inplace=True) # la colum country pase a ser el indice del dataframe
df_chi.head()
```

```python
df_chi.index.name = None #bajar un nivel
```

```python
df_chi.head(3)
```
